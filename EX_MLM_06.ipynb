```python
#Linearly Separable
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# No Overlap
X1, y1 = make_blobs(n_samples=90, centers=3, n_features=2, cluster_std=0.5, random_state=42)
clf1 = SVC(kernel='linear', C=1)
clf1.fit(X1, y1)

h = 0.02
x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max() + 1
y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z1 = clf1.predict(np.c_[xx.ravel(), yy.ravel()])
Z1 = Z1.reshape(xx.shape)

plt.contourf(xx, yy, Z1, alpha=0.3)
plt.scatter(X1[:, 0], X1[:, 1], c=y1, s=30, edgecolors='k')
plt.scatter(clf1.support_vectors_[:, 0], clf1.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='r')
plt.title("Linearly Separable - No Overlap")
plt.show()
```

```python
# Overlap
X2, y2 = make_blobs(n_samples=90, centers=3, n_features=2, cluster_std=1.5, random_state=42)
clf2 = SVC(kernel='linear', C=1)
clf2.fit(X2, y2)

x_min, x_max = X2[:, 0].min() - 1, X2[:, 0].max() + 1
y_min, y_max = X2[:, 1].min() - 1, X2[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z2 = clf2.predict(np.c_[xx.ravel(), yy.ravel()])
Z2 = Z2.reshape(xx.shape)

plt.contourf(xx, yy, Z2, alpha=0.3)
plt.scatter(X2[:, 0], X2[:, 1], c=y2, s=30, edgecolors='k')
plt.scatter(clf2.support_vectors_[:, 0], clf2.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='r')
plt.title("Linearly Separable - Overlap")
plt.show()
```

```python
#Non- Linearly Separable
from sklearn.datasets import make_circles

# No Overlap
X_a, y_a = make_circles(n_samples=60, factor=0.3, noise=0.02, random_state=42)
X_b, y_b = make_circles(n_samples=60, factor=0.6, noise=0.02, random_state=24)

y_b = np.where(y_b == 0, 2, 1)  
X = np.vstack([X_a, X_b])
y = np.hstack([y_a, y_b])

clf = SVC(kernel='rbf', C=100, gamma=0.5)
clf.fit(X, y)

h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, edgecolors='k')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='r')
plt.title("Non-linear Separable - No Overlap")
plt.show()
```

```python
# Overlap
X_a, y_a = make_circles(n_samples=60, factor=0.3, noise=0.1, random_state=42)
X_b, y_b = make_circles(n_samples=60, factor=0.6, noise=0.1, random_state=24)

y_b = np.where(y_b == 0, 2, 1)  
X2 = np.vstack([X_a, X_b])
y2 = np.hstack([y_a, y_b])

clf2 = SVC(kernel='rbf', C=1, gamma=0.5)
clf2.fit(X2, y2)

x_min, x_max = X2[:, 0].min() - 1, X2[:, 0].max() + 1
y_min, y_max = X2[:, 1].min() - 1, X2[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z2 = clf2.predict(np.c_[xx.ravel(), yy.ravel()])
Z2 = Z2.reshape(xx.shape)

plt.contourf(xx, yy, Z2, alpha=0.3)
plt.scatter(X2[:, 0], X2[:, 1], c=y2, s=30, edgecolors='k')
plt.scatter(clf2.support_vectors_[:, 0], clf2.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='r')
plt.title("Non-linear Separable - Overlap")
plt.show()
```

```python
#Non- separable
X, y = make_blobs(n_samples=90, centers=3, n_features=2, cluster_std=2.5, random_state=42)

clf = SVC(kernel='rbf', C=0.5, gamma=0.5)
clf.fit(X, y)

h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, edgecolors='k')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='r')
plt.title("Non-separable (Overlapping) Data")
plt.show()
```
